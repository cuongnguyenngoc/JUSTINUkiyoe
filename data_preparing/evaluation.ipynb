```python
import pandas as pd
```

```python
eval_1_path = 'Data/stage 3/Stage 3 Ukiyo-e Image Evaluation 1.csv'
eval_2_path = 'Data/stage 3/Stage 3 Ukiyo-e Image Evaluation 2.csv'
```

```python
df1 = pd.read_csv(eval_1_path)
df1.head()
```

```python
df2 = pd.read_csv(eval_2_path)
df2.head()
```

```python
total_by_keywords = 0
corrected_by_keywords = 0

for col in df1.columns:
    if 'keyword' in col:
        for img in df1[col]:
            total_by_keywords += 1
            if img == 'Image 1':
                corrected_by_keywords += 1
                
for col in df2.columns:
    if 'keyword' in col:
        for img in df2[col]:
            total_by_keywords += 1
            if img == 'Image 1':
                corrected_by_keywords += 1

print('Corrected answers based on keywords ', corrected_by_keywords)
print('Total answers based on keywords ', total_by_keywords)

print('Percentage of corrected answers based on keywords ', corrected_by_keywords/total_by_keywords)
```

```python
total_by_best_voting_description = 0
corrected_by_best_voting_description = 0
count = 1
for col in df1.columns:
    
    if count == 11:
        break
    if 'description' in col:
        for img in df1[col]:
            total_by_best_voting_description += 1
            if img == 'Image 1':
                corrected_by_best_voting_description += 1

        count += 1

count = 1
for col in df2.columns:
    if count == 11:
        break
    if 'description' in col:
        for img in df2[col]:
            total_by_best_voting_description += 1
            if img == 'Image 1':
                corrected_by_best_voting_description += 1

        count += 1
print()
print('Corrected answers based on best voting description ', corrected_by_best_voting_description)
print('Total answers based on best voting description ', total_by_best_voting_description)

print('Percentage of corrected answers based on best voting description ', \
      corrected_by_best_voting_description/total_by_best_voting_description)
```

```python
total_by_expert_description = 0
corrected_by_expert_description = 0
count = 1
for col in df1.columns:
    if count > 26:
        for img in df1[col]:
            total_by_expert_description += 1
            if img == 'Image 1':
                corrected_by_expert_description += 1

    count += 1
    
count = 1
for col in df2.columns:
    if count > 26:
        for img in df2[col]:
            total_by_expert_description += 1
            if img == 'Image 1':
                corrected_by_expert_description += 1

    count += 1
    
print()
print('Corrected answers based on expert description ', corrected_by_expert_description)
print('Total answers based on expert description ', total_by_expert_description)

print('Percentage of corrected answers based on expert description ', \
      corrected_by_expert_description/total_by_expert_description)
```
