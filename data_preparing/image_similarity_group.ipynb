```python
def split_keywords(x):
    keywords = []
    for key in x:
        keywords.extend(key.split('/'))
    return keywords
        
```

```python
df = pd.read_csv(ukiyo_content_file)

df = df[pd.notna(df['Keywords'])]
df.to_csv('Data/cleaned-ukiyo-e-content-remove-nan-keyword.csv')
df['keywords'] = df['Keywords'].apply(lambda x: x.split(';')).apply(lambda x: split_keywords(x))
for i, row in df.iterrows():
    j = 0
    for key in row['keywords']:
        df.at[i, 'keyword ' + str(j)] = re.sub('[^A-Za-z0-9]+', '', key)
        j += 1

df = df.groupby(['keyword 0']).size().reset_index(name='Count')
# df = df.groupby(['keyword 0'])['Source'].apply(lambda x: ';'.join(x)).reset_index(name='Sources')
# df.to_csv('Data/Similar-images.csv', index=None)
df
```

```python
import random, requests, os

df = pd.read_csv('Data/cleaned-ukiyo-e-content-remove-nan-keyword-filtered.csv')
# images = random.sample(list(df['Source']), 6)

# images

for url in df.Source:
    img_name = os.path.basename(url)
    if not os.path.exists('Data/images/' + img_name):
        img_data = requests.get(url).content
        with open('Data/images/' + img_name, 'wb') as handler:
            handler.write(img_data)
        print(url)
```

```python
import os

df = pd.read_csv('Data/cleaned-ukiyo-e-content-remove-nan-keyword-filtered.csv')

def check_url(url):
    img_name = os.path.basename(url)
    if os.path.exists('Data/Ukiyo-e Images/' + img_name):
        return True
    return False

df = df[df['Source'].apply(lambda x: check_url(x))]
df['Source'].to_csv('Data/Ukiyo-e Images/image-urls.csv', index=None)
        
```
